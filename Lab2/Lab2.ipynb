### Parentesi su condition number

I condition number di una matrice possono essere calcolati o tramite la definizione (usando l'inversa della matrice) o tramite altre funzioni come np.linalg.cond().
I risultati possono essere diversi quando la matrice è ill-conditioned, perchè se una matrice è ill-conditioned è quasi singolare e quindi ha il det circa = 0 e quindi "quasi non invertibile". Quindi calcolando l'inversa di una matrice ill-conditioned il risulato potrebbe essere diverso dalla vera inversa e dare quindi condition number diversi da quelli veri

# Da Wikipedia
## Matrices
For example, the condition number associated with the linear equation Ax = b gives a bound on how inaccurate the solution x will be after approximation. Note that this is before the effects of round-off error are taken into account; conditioning is a property of the matrix, not the algorithm or floating-point accuracy of the computer used to solve the corresponding system. In particular, one should think of the condition number as being (very roughly) the rate at which the solution x will change with respect to a change in b. Thus, if the condition number is large, even a small error in b may cause a large error in x. On the other hand, if the condition number is small, then the error in x will not be much bigger than the error in b.

The condition number is defined more precisely to be the maximum ratio of the relative error in x to the relative error in b.

Let e be the error in b. Assuming that A is a nonsingular matrix, the error in the solution A−1b is A−1e. The ratio of the relative error in the solution to the relative error in b is

{\displaystyle {\frac {\left\|A^{-1}e\right\|}{\left\|A^{-1}b\right\|}}/{\frac {\|e\|}{\|b\|}}={\frac {\left\|A^{-1}e\right\|}{\|e\|}}{\frac {\|b\|}{\left\|A^{-1}b\right\|}}.}{\displaystyle {\frac {\left\|A^{-1}e\right\|}{\left\|A^{-1}b\right\|}}/{\frac {\|e\|}{\|b\|}}={\frac {\left\|A^{-1}e\right\|}{\|e\|}}{\frac {\|b\|}{\left\|A^{-1}b\right\|}}.}
The maximum value (for nonzero b and e) is then seen to be the product of the two operator norms as follows:

{\displaystyle {\begin{aligned}\max _{e,b\neq 0}\left\{{\frac {\left\|A^{-1}e\right\|}{\|e\|}}{\frac {\|b\|}{\left\|A^{-1}b\right\|}}\right\}&=\max _{e\neq 0}\left\{{\frac {\left\|A^{-1}e\right\|}{\|e\|}}\right\}\,\max _{b\neq 0}\left\{{\frac {\|b\|}{\left\|A^{-1}b\right\|}}\right\}\\&=\max _{e\neq 0}\left\{{\frac {\left\|A^{-1}e\right\|}{\|e\|}}\right\}\,\max _{x\neq 0}\left\{{\frac {\|Ax\|}{\|x\|}}\right\}\\&=\left\|A^{-1}\right\|\,\|A\|.\end{aligned}}}{\displaystyle {\begin{aligned}\max _{e,b\neq 0}\left\{{\frac {\left\|A^{-1}e\right\|}{\|e\|}}{\frac {\|b\|}{\left\|A^{-1}b\right\|}}\right\}&=\max _{e\neq 0}\left\{{\frac {\left\|A^{-1}e\right\|}{\|e\|}}\right\}\,\max _{b\neq 0}\left\{{\frac {\|b\|}{\left\|A^{-1}b\right\|}}\right\}\\&=\max _{e\neq 0}\left\{{\frac {\left\|A^{-1}e\right\|}{\|e\|}}\right\}\,\max _{x\neq 0}\left\{{\frac {\|Ax\|}{\|x\|}}\right\}\\&=\left\|A^{-1}\right\|\,\|A\|.\end{aligned}}}
The same definition is used for any consistent norm, i.e. one that satisfies

{\displaystyle \kappa (A)=\left\|A^{-1}\right\|\,\left\|A\right\|\geq \left\|A^{-1}A\right\|=1.}{\displaystyle \kappa (A)=\left\|A^{-1}\right\|\,\left\|A\right\|\geq \left\|A^{-1}A\right\|=1.}
When the condition number is exactly one (which can only happen if A is a scalar multiple of a linear isometry), then a solution algorithm can find (in principle, meaning if the algorithm introduces no errors of its own) an approximation of the solution whose precision is no worse than that of the data.

However, it does not mean that the algorithm will converge rapidly to this solution, just that it will not diverge arbitrarily because of inaccuracy on the source data (backward error), provided that the forward error introduced by the algorithm does not diverge as well because of accumulating intermediate rounding errors.[clarification needed]

The condition number may also be infinite, but this implies that the problem is ill-posed (does not possess a unique, well-defined solution for each choice of data; that is, the matrix is not invertible), and no algorithm can be expected to reliably find a solution.

The definition of the condition number depends on the choice of norm, as can be illustrated by two examples.

If {\displaystyle \|\cdot \|}\|\cdot \| is the matrix norm induced by the (vector) Euclidean norm (sometimes known as the L2 norm and typically denoted as {\displaystyle \|\cdot \|_{2}}\|\cdot \|_{2}), then

{\displaystyle \kappa (A)={\frac {\sigma _{\text{max}}(A)}{\sigma _{\text{min}}(A)}},}{\displaystyle \kappa (A)={\frac {\sigma _{\text{max}}(A)}{\sigma _{\text{min}}(A)}},}
where {\displaystyle \sigma _{\text{max}}(A)}{\displaystyle \sigma _{\text{max}}(A)} and {\displaystyle \sigma _{\text{min}}(A)}{\displaystyle \sigma _{\text{min}}(A)} are maximal and minimal singular values of {\displaystyle A}A respectively. Hence:

If {\displaystyle A}A is normal, then
{\displaystyle \kappa (A)={\frac {\left|\lambda _{\text{max}}(A)\right|}{\left|\lambda _{\text{min}}(A)\right|}},}{\displaystyle \kappa (A)={\frac {\left|\lambda _{\text{max}}(A)\right|}{\left|\lambda _{\text{min}}(A)\right|}},}
where {\displaystyle \lambda _{\text{max}}(A)}{\displaystyle \lambda _{\text{max}}(A)} and {\displaystyle \lambda _{\text{min}}(A)}{\displaystyle \lambda _{\text{min}}(A)} are maximal and minimal (by moduli) eigenvalues of {\displaystyle A}A respectively.
If {\displaystyle A}A is unitary, then {\displaystyle \kappa (A)=1.}{\displaystyle \kappa (A)=1.}
The condition number with respect to L2 arises so often in numerical linear algebra that it is given a name, the condition number of a matrix.

If {\displaystyle \|\cdot \|}\|\cdot \| is the matrix norm induced by the {\displaystyle L^{\infty }}L^{\infty } (vector) norm] and {\displaystyle A}A is lower triangular non-singular (i.e. {\displaystyle a_{ii}\neq 0}{\displaystyle a_{ii}\neq 0} for all {\displaystyle i}i), then

{\displaystyle \kappa (A)\geq {\frac {\max _{i}{\big (}|a_{ii}|{\big )}}{\min _{i}{\big (}|a_{ii}|{\big )}}}}{\displaystyle \kappa (A)\geq {\frac {\max _{i}{\big (}|a_{ii}|{\big )}}{\min _{i}{\big (}|a_{ii}|{\big )}}}}
recalling that the eigenvalues of any triangular matrix are simply the diagonal entries.

The condition number computed with this norm is generally larger than the condition number computed relative to the Euclidean norm, but it can be evaluated more easily (and this is often the only practicably computable condition number, when the problem to solve involves a non-linear algebra[clarification needed], for example when approximating irrational and transcendental functions or numbers with numerical methods).

If the condition number is not too much larger than one, the matrix is well-conditioned, which means that its inverse can be computed with good accuracy. If the condition number is very large, then the matrix is said to be ill-conditioned. Practically, such a matrix is almost singular, and the computation of its inverse, or solution of a linear system of equations is prone to large numerical errors. A matrix that is not invertible has condition number equal to infinity

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
import scipy.io

data = scipy.io.loadmat('./data/MNIST.mat')

X = data['X']
I = data['I']
```

```python
X.shape
```

```python
test = X[:,16]
test = np.reshape(test, (16,16))
plt.imshow(test, cmap='gray')
```

```python
# Extract the subdataset of X that contains digits 3 and 4
digits = [3, 4]

_, indexes = np.where((I==digits[0]) | (I==digits[1]))

I = I[:, indexes]
X = X[:, indexes]

print(I.shape)
print(X.shape)

# Test
t=1
print(I[:,t])
plt.imshow(X[:,t].reshape(16,16), cmap='gray')
```

```python
# Separate training and test

from sklearn.model_selection import train_test_split

X_train, X_test, I_train, I_test = train_test_split(X.T, I.T, test_size=.2, random_state=42)
X_train, X_test, I_train, I_test = X_train.T, X_test.T, I_train.T, I_test.T


print(X_train.shape)
print(X_test.shape)
print(I_train.shape)
print(I_test.shape)
```

```python
# Extract digit-wise train matrices

_, index_3 = np.where(I_train==digits[0])
_, index_4 = np.where(I_train==digits[1])

X_1 = X_train[:, index_3]
X_2 = X_train[:, index_4]

print(X_1.shape)
print(X_2.shape)
```

```python
# Compute the SVD decomposition of X_1 and X_2

U_1, _, _ = np.linalg.svd(X_1, full_matrices=False)
U_2, _, _ = np.linalg.svd(X_2, full_matrices=False)
```

```python
# Take a new, unknown digit for the test set.

sample = 1
y = X_test[:, sample]
plt.imshow(X_test[:,sample].reshape(16,16), cmap='gray')

# Compute the projections of y into the two spaces
y_1 = U_1 @ (U_1.T @ y)
y_2 = U_2 @ (U_2.T @ y)
```

```python
# Compute the distances
d1 = np.linalg.norm(y_1-y)
d2 = np.linalg.norm(y_2-y)
```

```python
# Assign to the predicted class
if d1 < d2:
    predicted_class = digits[0]
else:
    predicted_class = digits[1]
    
# Print out
print('Predicted: ', predicted_class)
print('True: ', I_test[:, sample])

```

```python
def classify_2(d1, d2, y_test, classes):

    y_pred = [classes[0] if d1[i] < d2[i] else classes[1] for i in range(len(d1))]
    mse = np.mean(y_pred != y_test)*100
    nmse = np.sum(y_pred != y_test)

    return mse, nmse
        

def SVD_classification_2(X_train, y_train, X_test, y_test, classes):

    # Extract digit-wise train matrices
    _, index_1 = np.where(y_train==classes[0])
    _, index_2 = np.where(y_train==classes[1])

    X_1 = X_train[:, index_1]
    X_2 = X_train[:, index_2]

    # Compute the SVD decomposition of X_1 and X_2
    U_1, _, _ = np.linalg.svd(X_1, full_matrices=False)
    U_2, _, _ = np.linalg.svd(X_2, full_matrices=False)

    # Compute the projections of y into the two spaces
    y_1 = U_1 @ (U_1.T @ X_test)
    y_2 = U_2 @ (U_2.T @ X_test)

    d1 = np.linalg.norm(y_1-X_test, axis=0)
    d2 = np.linalg.norm(y_2-X_test, axis=0)

    mse, nmse = classify_2(d1, d2, y_test, classes)

    print(f'Misclassification error for digits {classes[0]} and {classes[1]}:')
    print(f'{mse:.2f}%')
    print()
    print(f'Number of misclassified digits:')
    print(nmse)
    
    return mse

def run_experiment_2(X, I, digits, err):

    # Extract the subdataset of X that contains desidered digits
    _, indexes = np.where((I==digits[0]) | (I==digits[1]))

    I = I[:, indexes]
    X = X[:, indexes]

    # Split train and test set 
    X_train, X_test, I_train, I_test = train_test_split(X.T, I.T, test_size=.2, random_state=42)
    X_train, X_test, I_train, I_test = X_train.T, X_test.T, I_train.T, I_test.T

    # Bulk classification
    mse = SVD_classification_2(X_train, I_train, X_test, I_test, digits)

    err[f'{digits[0]}_{digits[1]}'] = round(mse,2)

    return
    
```

```python
# Bulk classification

SVD_classification_2(X_train, I_train, X_test, I_test, digits)
```

```python
run_experiment_2(X, I, [5,1])
```

## 3-digits example

```python
def classify_3(d1, d2, d3, y_test, classes):

    y_pred = [classes[np.argmin([d1[i], d2[i], d3[i]])] for i in range(len(d1))]
    mse = np.mean(y_pred != y_test)*100
    nmse = np.sum(y_pred != y_test)

    return mse, nmse
        

def SVD_classification_3(X_train, y_train, X_test, y_test, classes):

    # Extract digit-wise train matrices
    _, index_1 = np.where(y_train==classes[0])
    _, index_2 = np.where(y_train==classes[1])
    _, index_3 = np.where(y_train==classes[2])

    X_1 = X_train[:, index_1]
    X_2 = X_train[:, index_2]
    X_3 = X_train[:, index_3]

    # Compute the SVD decomposition of X_1 and X_2
    U_1, _, _ = np.linalg.svd(X_1, full_matrices=False)
    U_2, _, _ = np.linalg.svd(X_2, full_matrices=False)
    U_3, _, _ = np.linalg.svd(X_3, full_matrices=False)

    # Compute the projections of y into the two spaces
    y_1 = U_1 @ (U_1.T @ X_test)
    y_2 = U_2 @ (U_2.T @ X_test)
    y_3 = U_3 @ (U_3.T @ X_test)

    d1 = np.linalg.norm(y_1-X_test, axis=0)
    d2 = np.linalg.norm(y_2-X_test, axis=0)
    d3 = np.linalg.norm(y_3-X_test, axis=0)

    mse, nmse = classify_3(d1, d2, d3, y_test, classes)

    print(f'Misclassification error for digits {classes[0]}, {classes[1]} and {classes[2]}:')
    print(f'{mse:.2f}%')
    print()
    print(f'Number of misclassified digits:')
    print(nmse)
    
    return mse

def run_experiment_3(X, I, digits, err):

    # Extract the subdataset of X that contains desidered digits
    _, indexes = np.where((I==digits[0]) | (I==digits[1]) | (I==digits[2]))

    I = I[:, indexes]
    X = X[:, indexes]

    # Split train and test set 
    X_train, X_test, I_train, I_test = train_test_split(X.T, I.T, test_size=.2, random_state=42)
    X_train, X_test, I_train, I_test = X_train.T, X_test.T, I_train.T, I_test.T

    # Bulk classification
    mse = SVD_classification_3(X_train, I_train, X_test, I_test, digits)

    err[f'{digits[0]}_{digits[1]}_{digits[2]}'] = round(mse,2)

    return
    
```
